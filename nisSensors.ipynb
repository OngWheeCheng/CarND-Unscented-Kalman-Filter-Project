```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nisLaser = np.loadtxt('./output/nisLaser.log')
nisRadar = np.loadtxt('./output/nisRadar.log')
```

```python
fig = plt.figure(figsize=(15, 4))

ax = fig.add_subplot(121)
ax.set_title('Laser NIS', fontsize=18)
ax.set_xlabel('Measurement Index')
ax.set_ylabel('Chi-squared')
ax.plot(nisLaser)
ax.plot((0, 250), (5.991, 5.991), 'r-', label = '5.991')
ax.legend(loc='upper right')
ax.axis([0, 250, 0, 18])

ax = fig.add_subplot(122)
ax.set_title('Radar NIS', fontsize=18)
ax.set_xlabel('Measurement Index')
ax.set_ylabel('Chi-squared')
ax.plot(nisRadar)
ax.plot((0, 250), (7.815, 7.815), 'r-', label = '7.815')
ax.legend(loc='upper right')
ax.axis([0, 250 , 0, 18])

plt.savefig('./output/NIS.png')
plt.show()
```

```python
count = 0
for num in range(len(nisLaser)):
    if (nisLaser[num] > 5.991):
        count = count + 1
print("{0:.2f}% of Laser NIS exceeds 5.991".format(count/len(nisLaser)*100))
```

```python
count = 0
for num in range(len(nisRadar)):
    if (nisLaser[num] > 7.815):
        count = count + 1
print("{0:.2f}% of Radar NIS exceeds 7.881\n".format(count/len(nisRadar)*100))
```
